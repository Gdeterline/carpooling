# Projet CASSIOPEE

# Description : 

The project aims to develop an algorithm that generates a weekly carpooling schedule for a group of parents based on their availability and their children's transportation needs. The algorithm should balance the number of rides requested for a child with the number of available seats in a parent's car, to maximize the number of users satisfied and the car occupancy rate. The project may also include building a user-friendly interface for collecting user requests and exploring market opportunities for commercializing the solution. The goal is to reduce greenhouse gas emissions by promoting carpooling as an alternative to individual car transportation.

### Assumption 1
Each passenger issues one request per day and per direction

### Simplification 1


INPUT :

$$
\mathcal{O} = \{ \mathcal{o_1} , \mathcal{o_2}, \mathcal{o_3}, ... \} \ \ \forall i \ \ \mathcal{o_i} = (d, t, wp, wi, p)  \\
With : \ \ d \rightarrow driver,\ \ t \rightarrow time, \ \ wp \rightarrow \ \ 1\ \ if \ \ week \ \ pair, \ \  wi \rightarrow \ \ 1 \ \ if \ \ week \ \ impair, \ \ p \rightarrow place \\ 
\mathcal{O}  \ \ represents \ \ all \ \ offers \ \ proposed \ \ by \ \ the \ \ drivers \\
\\ \ \ \\
\mathcal{R} = \{ \mathcal{r_1} , \mathcal{r_2}, \mathcal{r_3}, ... \} \ \ \forall i \ \ \mathcal{r_i} = (p, t, wp, wi)  \\ 
With : \ \ p \rightarrow passager,\ \ t \rightarrow time, \ \ wp \rightarrow \ \ 1\ \ if \ \ week \ \ pair, \ \  wi \rightarrow \ \ 1 \ \ if \ \ week \ \ impair, \ \ \\
\mathcal{R} \ \ is \ \ the \ \ set \ \ of \ \ requests \ \ from \ \  passengers
\\ \ \ \\
A_{c, t, w} = \{ 0 , 1 \} \ \ \rightarrow 1 \ \ if \ \ the \ \ driver \ \ c \ \ is \ \ for \ \ time \ \ t \ \ and \ \ week \ \ w \\
B_{n, t, w} = \{ 0 , 1 \} \ \ \rightarrow 1 \ \ if \ \ the \ \ passager \ \ n \ \ is \ \ for  \ \ time \ \ t \ \ and \ \ week \ \  w \\
M_{n, c, t, w} = \{ 0 , 1 \} \ \ \rightarrow 1 \ \ if \ \ driver \ \ c \ \ is \ \ parent \ \ of \ \ the \ \  child \ \ n  \\
K_{c}  \rightarrow number \ \ of \ \ places \ \ in \ \ the \ \ car \ \ of \ \ the \ \ driver \ \ of \ \  the \ \ offer \ \ o  \\
$$

## Decision Variables :

$$
X_{n, c, t, w} = \{ 0 , 1 \} \ \ \rightarrow 1 \ \ if \ \ we \ \ accept \ \ the \ \ offer \ \ o \ \ for \ \ the \ \ request \ \ r \ \ on \ \ the \ \ time \ \ t \ \ and \ \ the \ \ week \ \ w   \\
E_{c, t, w} = \{ 0 , 1 \} \ \ \rightarrow 1 \ \ if \ \ the \ \ driver \ \ c \ \ is \ \ request \ \ for \ \ the \ \ time \ \ t \ \ and \ \ the \ \ week \ \ w   \\
N_{c} \ \ type \ \ Int \ \ represent \ \ the \ \ number \ \ of \ \ car \ \ used \ \ for \ \ the \ \ time \ \ t \ \ and \ \ the \ \ week \ \ w   \\
$$

## The constraints :
$$
Each \ \ request \ \ can \ \ only \ \ be \ \ attributed \ \ to \ \ one \ \ offer \ \ at \ \ a \ \ time \ \ t \ \ and \ \ a \ \ week \ \ w \\
\forall n, t, w \sum_{o} X_{n, c, t, w} A_{n, t, w} B_{c, t, w} \le 1 \\
Number \ \ of \ \ request \ \ attribute \ \ to \ \ each \ \ offer \ \ o \ \ must \ \ be \ \ less \ \ than \ \ the \ \ number \ \ of \ \ places \ \ available : \\
\forall c, t, w \sum_{n} X_{n, c, t, w} \le K_o \\
\\
No \ \ request \ \ can \ \ be \ \ attributed \ \ to \ \ a \ \ non-existent \ \ offer : \\
if \ \ A_{c, t, w}==0, \ \ \sum_{n} X_{n, c, t, w} == 0 \\
No \ \ offer \ \ can \ \ be \ \ associated \ \ to \ \ a \ \ non-existent \ \ request : \\
if \ \ B_{n, t, w}==0, \ \ \sum_{c} X_{n, c, t, w} == 0 \\
E_{c,t,w} > 10^{-6} * \sum_{n, t, w} X_{n, c, t, w} \\
G_c = \sum_{t, w} E_{c, t, w}, \ \ \forall c
$$

### Objective Fonction :

The objective function aims to maximise the number of satisfied requests :
$$
max \sum_{n, c, t, w} X_{n,c,t,w} .* M_{n, c}(\beta) - \alpha \sum_c G_c
$$

```julia
# Déterminer le nombre d'offres

N_o = sum(x -> x in ["Oui", "Pair", "Impair"], data_offers)

# Initialiser la matrice O
O = zeros(Int, N_o, 5)

# Compteur pour les offres valides
idx = 1

# Parcourir le tableau offers
for i in 1:size(K, 1)
    nb_places = K[i]
    for j in 1:size(data_offers, 2)
        if data_offers[i, j] == "Oui"
            # Stocker l'offre dans la matrice O
            O[idx, 1] = i
            O[idx, 2] = j
            O[idx, 3] = 1
            O[idx, 4] = 1
            O[idx, 5] = nb_places
            idx += 1
        end
        if data_offers[i, j] == "Pair"
            # Stocker l'offre dans la matrice O
            O[idx, 1] = i
            O[idx, 2] = j
            O[idx, 3] = 1
            O[idx, 4] = 0
            O[idx, 5] = nb_places
            idx += 1
        end
        if data_offers[i, j] == "Impair"
            # Stocker l'offre dans la matrice O
            O[idx, 1] = i
            O[idx, 2] = j
            O[idx, 3] = 0
            O[idx, 4] = 1
            O[idx, 5] = nb_places
            idx += 1
        end
    end
end


println("Matrice O :")
println(O)
println()

# Déterminer le nombre de requêtes

N_r = sum(x -> x in ["Oui", "Pair", "Impair"], data_requests)

# Initialiser la matrice R
R = zeros(Int, N_r, 4)

# Compteur pour les offres valides
idr = 1

# Parcourir le tableau offers
for i in 1:size(data_requests, 1)
    for j in 1:size(data_requests, 2)
        if data_requests[i, j] == "Oui"
            # Stocker l'offre dans la matrice R
            R[idr, 1] = i
            R[idr, 2] = j
            R[idr, 3] = 1
            R[idr, 4] = 1
            idr += 1
        end
        if data_requests[i, j] == "Pair"
            # Stocker l'offre dans la matrice R
            R[idr, 1] = i
            R[idr, 2] = j
            R[idr, 3] = 1
            R[idr, 4] = 0
            idr += 1
        end
        if data_requests[i, j] == "Impair"
            # Stocker l'offre dans la matrice R
            R[idr, 1] = i
            R[idr, 2] = j
            R[idr, 3] = 0
            R[idr, 4] = 1
            idr += 1
        end
    end
end

println("Matrice R :")
println(R)

#Afficher les drivers et les passagers
println()
# Parcourir tous les jours et toutes les heures

for t in 1:T
    # Trouver les conducteurs disponibles à l'heure et les semaines pairs
    available_drivers_pairs = unique(O[(O[:, 2] .== t) .& (O[:,3] .== 1), 1])
    # Trouver les conducteurs disponibles à l'heure et les semaines imapirs
    available_drivers_impairs = unique(O[(O[:, 2] .== t) .& (O[:,4] .== 1), 1])

    # Afficher les conducteurs disponibles à l'heure et les semaines pairs
    println("Driver disponible pour une conduite à l'horaire $t les semaines pairs :")
    println(available_drivers_pairs)
    
    # Afficher les conducteurs disponibles à l'heure et les semaines impairs
    println("Driver disponible pour une conduite à l'horaire $t les semaines impairs :")
    println(available_drivers_impairs)
end

println()

for t in 1:T
    # Trouver les conducteurs disponibles à l'heure t et les semaines pairs
    available_passagers_pairs = unique(R[(R[:, 2] .== t) .& (R[:, 3] .== 1), 1])
    # Trouver les conducteurs disponibles à l'heure t et les semaines impairs
    available_passagers_impairs = unique(R[(R[:, 2] .== t) .& (R[:, 4] .== 1), 1])

    # Afficher les conducteurs disponibles à l'heure et au jour donné les semaines pairs
    println("Passager demandant une conduite à l'horaire $t les semaines pairs:")
    println(available_passagers_pairs)
    
    # Afficher les conducteurs disponibles à l'heure et au jour donné les semaines impairs
    println("Passager demandant une conduite à l'horaire $t les semaines impairs:")
    println(available_passagers_impairs)
end
```

```julia
function covoiturage(Beta, Alpha, time)
    
    M = zeros(Float64, N, C, T, 2)
    
    for parent in 1:size(name_offers,1)
        for child in 1:size(name_requests,1)
            if name_requests[child] == child_offers[parent]
                for t in 1:T
                    for w in 1:2
                        M[child, parent, t, w] = Beta
                    end
                end
            else
                for t in 1:T
                    for w in 1:2
                        M[child, parent, t, w] = 10-Beta
                    end
                end
            end
        end
    end

    # Création du modèle
    m = Model(GLPK.Optimizer)


    # Variables de décision
    @variable(m, X[1:N, 1:C, 1:T, 1:2], Bin) #Cette variable binaire représente si un enfant i est affecté à une voiture j.
    @variable(m, E[1:C, 1:T, 1:2], Bin)
    @variable(m, G[1:C], Int)

    # Contraintes

    for t in 1:T
        for w in 1:2
            for c in 1:C
                @constraint(m, E[c,t,w] >= 10^(-2)*sum(X[n, c, t, w] for n in 1:N))
            end
        end
    end

    #nombre de trajet effectué par c

    for c in 1:C
        @constraint(m, sum(E[c,t, w] for t in 1:T, w in 1:2) == G[c] )
    end

    #nombre de trajet

    for n in 1:N
        @constraint(m, sum(X[n, c, t, w] for c in 1:C, t in 1:T, w in 1:2 ) <= count_passager[n,2])
    end

    # La capacité maximale de chaque voiture doit être respectée.

    for t in 1:T
        for c in 1:C
            for w in 1:2
                @constraint(m, sum(X[n, c, t, w] for n in 1:N) <= K[c])
            end
        end
    end


    # Un enfant ne peut être transporté que dans une seule voiture.

    for t in 1:T
        for n in 1:N
            for w in 1:2
                @constraint(m, sum(X[n, c, t, w]*A[c, t, w]*B[n, t, w] for c in 1:C) <= 1)
            end
        end
    end


    for t in 1:T
        for c in 1:C
            for w in 1:2
                if A[c, t, w] == 0
                    @constraint(m, sum(X[n, c, t, w] for n in 1:N) == 0)
                end
            end
        end
    end



    for t in 1:T
        for n in 1:N
            for w in 1:2
                if B[n, t, w] == 0
                     @constraint(m, sum(X[n, c, t, w] for c in 1:C) == 0)
                end
            end
        end
    end


    # Fonction objectif
    
    @objective(m, Max, sum(X .* M) - Alpha*sum(G))
 
    

    # Résolution du problème
    MOI.set(m, MOI.TimeLimitSec(), time)
    
    optimize!(m)
    
    status = termination_status(m)

    if status == MOI.OPTIMAL
        # Parfait !
        println("Solution optimale trouvée:")
        # Récupérez et affichez la solution ici (par exemple, en utilisant la fonction value() pour les variables)
    elseif status == MOI.INFEASIBLE
        # Il n'existe aucune solution
        println("Il n'existe aucune solution :'(")
    elseif status == MOI.TIME_LIMIT && primal_status(m) == MOI.FEASIBLE_POINT
        # Solution réalisable trouvée, mais pas nécessairement optimale
        println("Solution réalisable trouvée dans la limite de temps:")
        # Récupérez et affichez la solution ici
    else
        # Voir la documentation
        println("Voir la documentation")
    end

    ########################################################################################################################
    #Compter le nombre de requête satisfaite
    ########################################################################################################################
    
    nb_request_done = 0
    
    for w in 1:2
        for t in 1:T
            for c in 1:C
                for n in 1:N
                    if value(X[n,c,t,w]) == 1
                        nb_request_done += 1
                    end
                end
            end
        end
    end
    
    pourcent_request = nb_request_done/nb_request*100
    
    println()
    println("Le nombre de demande de conduite faite par les enfants est : ", nb_request)
    println("Le nombre de demande de conduite satisfaite est : ", nb_request_done)
    println("Le pourcentage de demande satistaite est : ", pourcent_request , "%")
    println()
    
    ########################################################################################################################
    #Compter le taux de remplissage des voitures
    ########################################################################################################################
    
    nb_place_dispo = 0
    for w in 1:2
        for t in 1:T
            for c in 1:C
                res = 0
                for n in 1:N
                    if value(X[n,c,t,w]) == 1
                        res += 1
                    end
                    
                end
                if res >= 1
                    nb_place_dispo += number_places[c]
                end
            end
        end
    end
    
    pourcent_remplissage = nb_request_done/nb_place_dispo*100
    
    println("Nombre de place disponible après répartition : ", nb_place_dispo)
    println("Le taux de remplissage des voitures est : ", pourcent_remplissage, "%")
    println()

    ########################################################################################################################
    # Nombre d'enfant sans leur parent
    ########################################################################################################################
    
    I=0
    L=0
    E=0
    for w in 1:2
        for t in 1:T
            for c in 1:C
                J=0
                for n in 1:N
                    L += value(X[n, c, t, w])
                    J += value(X[n, c, t, w])
                    if value(M[n,c,t,w]) >= 5.01 && value(X[n, c, t, w]) == 1
                        E += 1
                    end
                end

                if J >= 1
                    I += 1
                end
            end
        end
    end
    
    println("Nombre d'enfants avec leur parent : ", E)
    println()
    
    pourcent_family = E/I*100
    
    println("Pourcentage d'enfant dans la voiture de leur parent : ", pourcent_family, " %")
    println()
    
    println("Nombre de voiture utilisée : ", I,", Nombre d'enfant dans les voitures : ",L,", Nombre d'enfants sans leur parent : ", E)
    println()
    
    ########################################################################################################################
    # Créer un tableau de résultat
    ########################################################################################################################
                    
    W = fill("/////////////", 300, T)

    idx = 0
    for i in 1:T
        W[1, i] = horaire_requests[i]
    end

    for t in 1:T
        h = horaire_requests[t]
        Time = "$h"
        println()
        println(Time)
        VAR1 = []
        for j in 1:C
            nb = K[j]
            d = name_offers[j] 
            e = number_places[j]
            f = d * " " * string(e) * " places"
            global offer = ""
            for w in 1:2
                VAR2 = []
                VAR3 = []
                if w==1
                    week = "Semaine paire"
                end
                if w==2
                    week = "Semaine Impaire"
                end
                push!(VAR2, week)
                push!(VAR2, f)
                push!(VAR3, week)
                push!(VAR3, f)
                offer = "$d with $nb places, week $week :"

                for i in 1:N
                    n = name_requests[i]
                    if value(X[i,j,t,w]) == 1
                        offer *= "$n, "
                        push!(VAR2, "$n")
                    end
                end
                if offer != "$d with $nb places, week $week :"
                    println(offer)
                end
                if VAR2 != VAR3
                    len = length(VAR2)
                    VAR1 = vcat(VAR1, VAR2)
                    Q = 15 - len
                    for f in 1:Q
                        push!(VAR1, "/")
                    end
                end
            end
        end
        for g in 1:size(VAR1,1)
            W[g+1, t] = VAR1[g]
        end
    end

    println()
    
    # Affichage du nombre de trajets par enfant
    
    # Initialisation du tableau des nombres de trajets par enfant
    nombre_trajets_par_enfant = zeros(Int, N)

    # Parcours des indices n, c, t, w
    for n in 1:N
        for c in 1:C
            for t in 1:T
                for w in 1:2
                    # Vérification si X[n, c, t, w] est égal à 1
                    if value(X[n, c, t, w]) == 1.0
                        # Incrémentation du nombre de trajets pour l'enfant n
                        nombre_trajets_par_enfant[n] += 1
                    end
                end
            end
        end
    end
    for n in 1:N
        nbu = count_passager[n,2]
        println("Enfant $n qui a le droit à $nbu trajets : $(nombre_trajets_par_enfant[n]) trajets")
    end

    println(count_passager)

    
    
    # Créer un fichier Excel avec le tableau W
    XLSX.openxlsx("Repartition_Voiture.xlsx", mode="w") do xf
    sheet = xf[1]
    XLSX.rename!(sheet, "new_sheet")
    
    sheet["A1:AA300"] = W
    end
end
```
